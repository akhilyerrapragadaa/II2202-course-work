```python
from PIL import Image
import pytesseract
import pdf2image
import time
```

```python
pages = pdf2image.convert_from_path(pdf_path='./Downloads/instructions-for-adding-your-logo.pdf', dpi=300)

start_time = time.time()
for i in range(len(pages)):
        content = pytesseract.image_to_string(pages[i], lang='eng')
        print(content)
print("--- %s seconds ---" % (time.time() - start_time))
```
